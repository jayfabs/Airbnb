```python
import geopandas as gpd
from shapely.geometry import Point
```

```python

# Load the Airbnb rental data into a Pandas dataframe

airbnb_df = pd.read_csv('data_clean_bnb.csv')
```

```python
# Create a GeoDataFrame by converting the latitude and longitude columns into Point objects
geometry = [Point(xy) for xy in zip(airbnb_df['longitude'], airbnb_df['latitude'])]
crs = {'init': 'epsg:4326'}
airbnb_gdf = gpd.GeoDataFrame(airbnb_df, crs=crs, geometry=geometry)
```

```python
# Convert the GeoDataFrame into GeoJSON format
airbnb_geojson = airbnb_gdf.to_crs(epsg='4326').to_json()
```

```python
# Save the GeoJSON data to a file
with open('airbnb_melbourne.geojson', 'w') as f:
    f.write(airbnb_geojson)
```
